```python
data['Compound'] = data['Compound'].fillna(data['Compound'].mode()[0])
```

```python
data.fillna(data.median(numeric_only=True), inplace=True)
```

```python
data['TyreWearRate'] = data['TyreLife'] / (data['LapNumber'] + 1)

data['LapSpeedTrend'] = data['LapTime (s)'] / data['RaceProgress']

data['PositionPressure'] = data['Position'] * data['RaceProgress']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['Compound'] = le.fit_transform(data['Compound'])
```

```python
x=data.drop(columns=['PitNextLap'])
y=data['PitNextLap']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 42)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE()

x_train, y_train = sm.fit_resample(x_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```
